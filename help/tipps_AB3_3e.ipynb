# Tipp für die Aufgabe 3e) des Arbeitsblattes 3

Es können unterschiedliche Möglichkeiten genutzt werden, um die Qubits über die Farbkanäle hinweg miteinander zu verschränken.

## Grundlegende Verschränkungs-Techniken

### 1) Bell-Paar mit 2 Qubits

1. Das Hadamard Gate auf dem ersten Qubit erzeugt eine Superposition.
2. Durch ein kontrolliertes X Gate (CX oder auch CNOT) mit dem ersten Qubit als Kontrolle und dem zweiten Qubit als Ziel, können die beiden Zustände gekoppelt werden.

````python
bell_circuit = QuantumCircuit(2)

bell_circuit.h(0)
bell_circuit.cx(0, 1)

bell_circuit.draw("mpl")
````

### 2) GHZ-Zustand mit mehreren Qubits

1. Das Hadamard Gate auf dem ersten Qubit erzeugt eine Superposition.
2. Der Einsatz von CNOT Gates von diesem Qubit zu allen anderen Qubits, die du verknüpfen willst (z.B. den entsprechenden Bits der Grün- und Blau-Qubits).

````python
ghz_circuit = QuantumCircuit(3)

ghz_circuit.h(0)
ghz_circuit.cx(0, 1)
ghz_circuit.cx(0, 2)

ghz_circuit.draw("mpl")
````

### 3) Verschränkung über ganze Blöcke

- Wiederholung des Bell-Musters für jedes Bit-Paar, z.B. (`R[i]`, `G[i]`) und anschließend (`R[i]`, `B[i]`)
- Verknüpfung aller Farbkanäle mit mehreren CNOTs, z.B. `CNOT(R[0], G[0])`, `CNOT(G[0], B[0])`, `CNOT(B[0], R[0])` usw.

````python
block_bell_circuit = QuantumCircuit(
    QuantumRegister(1, "Red"), 
    QuantumRegister(1, "Green"), 
    QuantumRegister(1, "Blue"))

block_bell_circuit.h(0)
block_bell_circuit.cx(0, 1)
block_bell_circuit.h(1)
block_bell_circuit.cx(1, 2)

block_bell_circuit.draw("mpl")
````

````python
cnot_circuit = QuantumCircuit(
    QuantumRegister(1, "Red"), 
    QuantumRegister(1, "Green"), 
    QuantumRegister(1, "Blue"))

cnot_circuit.cx(0, 1)
cnot_circuit.cx(1, 2)
cnot_circuit.cx(2, 0)

cnot_circuit.draw("mpl")
````
