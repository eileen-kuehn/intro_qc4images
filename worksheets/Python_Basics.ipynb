## 3) Funktionen

Zusammen mit Variablen sind Funktionen die Grundlage des Programmierens. Auch die kennst du wahrscheinlich schon aus dem Matheunterricht. Ein Beispiel wäre $y = 3x + 5$. In höheren Klassenstufen schreibt man beispielsweise auch $f(x) = 3x + 5$.

Das zeigt bereits ganz gut, wie es beim Programmieren ist. Eine Funktion, die `f` heißt, bekommt einen Wert (oder genauer gesagt eine Variable) `x` übergeben und rechnet mit dieser, um das Ergebnis zu bestimmen, z.B. $f(2) = 3 * 2 + 5 = 11$ für $x = 2$. 

Für die Rückgabe des Ergebnisses wird das Wort `return` verwendet. Das muss nicht zwingend passieren. Manchmal macht eine Funktion einfach nur etwas und gibt nichts zurück. Wie viele Variablen die Funktion bekommt, kann auch komplett unterschiedlich sein - auch keine Variablen sind möglich.  

Wie dir jetzt vielleicht auffällt, kanntest du von Anfang an schon eine Funktion, nämlich `print()`. Diese bietet uns Python selbst schon an. Diese gibt nichts zurück, sondern sie macht einfach nur etwas, indem sie die Variablen ausgibt, mit der man sie aufgerufen hat.

Aber jetzt kommt erstmal ein Beispiel. Versuche zu verstehen, was passiert. (Führe die Zelle aus, es wird aber nichts (sichtbares) passieren.) 

````python
# Funktionen beginnen in python immer mit "def" - am Ende ist ein Doppelpunkt
def add(a, b):
    result = a + b
    return result
````

Hier wurde eine Funktion `add` geschrieben, die zwei Variablen aufaddiert und das Ergebnis zurückgibt. 

Die beiden Variablen werden in der Funktion `a` und `b` genannt (so werden sie in der ersten Zeile genannt, der quasi die Namen so definiert). Wie die Variablen heißen, mit der wir die Funktion aufrufen, ist egal. In der Funktion werden die beiden Werte addiert und dann wird das Ergebnis mit `return` zurückgegeben. 

(Wie du siehst, benutzt man beim Programmieren eigentlich immer englische Namen für Funktionen (und auch Variablen), damit alle Menschen deinen Code schnell verstehen können. Deswegen haben wir sie `add()` genannt und nicht `addieren()`. Man kann sie aber eigentlich nennen, wie man will.)

Bei Funktionen in Python ist es wichtig, dass ab der Definition in der ersten Zeile alles eingerückt wird, bis die Funktionen endet. Das macht man entweder mit `4 Leerzeichen` oder (bevorzugt) indem man einmal `Tab` drückt.

Aber jetzt wollen wir auch mal unsere erste eigene Funktion nutzen. Führe die folgende Zelle mal aus. Die Funktion `add()`, die wir oben selbst geschrieben haben, können wir genauso nutzen, wie wir zB. `print()` genutzt haben. 

````python
var1 = 3
var2 = 9.9

# jetzt kommt der Funktionsaufruf
result = add(var1, var2)
print(result)
````

Vergiss nicht, `var1` und `var2` sind nur Platzhalter. In der Funktion wird `var1` `a` genannt. `var2` wird zu `b`.

Man kann auch die Funktion `f` erstellen, die genauso das macht, was wir als Mathe-Beispiel oben hatten.

````python
def f(x):
    y = 3 * x + 5
    return y


result = f(2)
print(result)
````

## 5) Listen

In Python kann man Listen definieren, die mehrere Variablen beinhalten. Listen werden mit eckigen Klammern gekennzeichnet. 

````python
a = 5
b = "Hello"
c = True

my_list = [a, b, c]

print(my_list)
````

Listen können auch leer sein. Mit `append()` kann man Dinge hinzufügen. Anders als bislang, ruft man diese Funktion auf, indem man den Aufruf an den Listenamen anhängt:

````python
# Wir definieren eine leere Liste
my_list = []

print(my_list)

# Wir fügen zwei Elemente hinzu
a = 6
my_list.append(a)
my_list.append(5)

print(my_list)
````

Man kann auch gezielt auf die Position in einer Liste zugreifen und diese auch so verändern. Das macht man mit den eckigen Klammern. Die gewünschte Position ist dazwischen.

<div id="information" class="alert alert-warning">
  <h2><i class="fas fa-info" style="font-size:36px"></i> </h2>
  <p>In der Informatik fangen wir bei <code>0</code> an zu zählen. Deswegen ist das erste Listenelement an Position <code>0</code>. Damit ist quasi jede Stelle um eins verschoben. Wenn eine Liste also <code>4</code> Elemente hätte, müssen wir auf das letzte Element mit <code>[3]</code> zugreifen.</p>

Mit den eckigen Klammern können wir auch Werte verändern. Hier ersetzen wir den `Dachs` durch einen `Dackel`.

````python
my_list[3] = "Dackel"
print(my_list)
````

`len()` ist eine Funktion, die uns ähnlich wie `print()` von Python zur Verfügung gestellt wird. Diese gibt uns die Länge einer Liste zurück oder besser gesagt: wie viele Elemente sind in der Liste. 

````python
print(len(my_list))
````

## 6) For-Schleifen

`For`-Schleifen sind neben `if`-Abfragen, Funktionen und Variablen ein weiteres Konzept, das man in fast jeder Programmiersprache findet. Mit ihnen kann man über Dinge "iterieren" (also "drübergehen") - in unserem Fall also Listen. Eine Schleife führt dabei eine eigene Variable ein (die man auch wieder beliebig benennen kann). Diese Variable nimmt dann jeden Wert einer Liste einmal an und mit diesem kann man dann etwas machen. 

Man kann deswegen folgende Schleife quasi lesen als: `Für jedes Element der Liste "numbers", das ich in dem Moment "n" nenne, mache das Folgende: print(n)`. 

````python
numbers = [1, 2, 3, 4, 5]

for n in numbers:
    print(n)
print("Liste zu Ende")
````

Wie du siehst muss der Code, der für jedes Element ausgeführt werden soll, auch wieder eingerückt sein.

Man kann auch kompliziertere Dinge mit den Elementen machen als sie einfach auszugeben. Was aber so nicht funktioniert, ist, die Liste zu verändern. Denn `n` ist ja einfach nur ein Name, wie wir das aktuelle Element nennen, aber nicht dessen Wert. 

Die folgende Zelle hat also keinen Effekt. `numbers` sieht danach genauso aus wie vorher.

````python
for n in numbers:
    n = n + 1

print(numbers)
````

Listen durch Schleifen zu verändern, geht aber schon. Da müssen wir aber anders drübergehen. Du hast ja weiter oben schon gelernt, dass man den Eintrag eine Liste ändern kann, wenn man dessen Position kennt (das war das mit den eckigen Klammern `[]`). Wenn wir also über eine Liste drübergehen würden, die alle möglichen Positionen unserer Liste enthält, könnten wir so auch die Inhalte der Liste bearbeiten.

Um das zu tun, können wir wieder eine Funktion verwenden, die uns Python bereitstellt und zwar `range()`. Der Aufruf `range(5)` gibt uns nämlich folgende Liste zurück `[0, 1, 2, 3, 4]`.

````python
for i in range(5):
    print(i)
````

Auch hier wird also von `0` an gezählt. Aber das ist ja sehr praktisch, wenn wir so unsere Liste verändern wollen. Denn für eine Liste mit `5` Elementen, kriegen wir mit `range(5)` alle möglichen Positionen. 
Und wenn wir nicht wissen, wie viele Elemente eine Liste hat, können wir das mit `len()` abfragen.  

<div id="exercise" class="alert alert-info">
  <h3><i class="fa fa-laptop" style="font-size:28px"></i> <br><br> a) Was wird ausgegeben? </h3>
  <p>Das folgende Stück Code macht vieles auf einmal, was du gerade gelernt hast. Guck mal, ob du vorhersagen kannst, was am Ende rauskommt.</p>

  ```python
numbers = [0, 1, 2, 3, 4]

length = len(numbers)

for i in range(length):
    numbers[i] = numbers[i] + 1

print(numbers)
```
